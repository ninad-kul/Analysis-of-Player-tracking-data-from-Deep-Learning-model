Functions for Centroid and Nearby players

```python
track_csv = '/content/contentdrive/MyDrive/Spordia/track_data.csv'    #Input csv address

def get_centroid(t,l,w,h):
    th = (t+h)/2
    wl = (w+l)/2
    return [th,wl]


def players_nearby(csv_address,frame_no,player_no):
    track_df = pd.read_csv(csv_address)
    nearby_ids = []
    frame_df = track_df.loc[track_df['frame_idx'] == frame_no]

    t = frame_df.loc[frame_df['track_id'] == player_no,'t'].iloc[0]
    l = frame_df.loc[frame_df['track_id'] == player_no,'l'].iloc[0]
    w = frame_df.loc[frame_df['track_id'] == player_no,'w'].iloc[0]
    h = frame_df.loc[frame_df['track_id'] == player_no,'h'].iloc[0]
    player_centrd = get_centroid(t,l,w,h)

    unique_tracks = frame_df['track_id'].unique()

    for i in unique_tracks:
        tc = frame_df.loc[frame_df['track_id'] == i,'t'].iloc[0]
        lc = frame_df.loc[frame_df['track_id'] == i,'l'].iloc[0]
        wc = frame_df.loc[frame_df['track_id'] == i,'w'].iloc[0]
        hc = frame_df.loc[frame_df['track_id'] == i,'h'].iloc[0]

        centrd_temp = get_centroid(tc,lc,wc,hc)
        dist_temp = math.dist(player_centrd,centrd_temp)
        tk_id = frame_df.loc[frame_df['track_id'] == i,'track_id'].iloc[0]
        print(tk_id," -- ",dist_temp)

        if dist_temp != 0 and dist_temp<100:
           nearby_ids.append(tk_id)
        else:
           pass
    return nearby_ids
print(frame_df)

players_nearby(track_data,202,7)

#get_centroid(878,383,98,232)
```
